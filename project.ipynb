# Data Analysis and Visualization

```python
import plotly.express as px
temp_df = df.copy()

columns_to_include = ['Age', 'CGPA', 'Have you ever had suicidal thoughts ?',
                      'Work/Study Hours', 'Financial Stress',
                      'Family History of Mental Illness', 'Depression',
                      'Overall Pressure', 'Overall Satisfaction']

temp_df = temp_df[columns_to_include]
correlation_matrix = temp_df.corr(method='pearson')

fig = px.imshow(
    correlation_matrix,
    text_auto=True,
    color_continuous_scale='RdBu',
    range_color=[-1, 1],
    labels={'color': "Pearson's Correlation"},
    x=correlation_matrix.columns,
    y=correlation_matrix.index
)

fig.update_layout(
    title={
        'text': "Correlation Matrix (with Gender)",
        'x': 0.01,
        'y': 0.97
    },
    paper_bgcolor="white",
    plot_bgcolor="white",
    xaxis_title="Features",
    yaxis_title="Features",
    coloraxis_colorbar_title="Pearson's Correlation",
    width=1200,
    height=900
)

fig.show()
```

To conclude, from this graph we can say that older age groups may have slightly lower levels of depression than younger ones. A value of -0.2 indicates a weak negative correlation.
This means that there is a link between age and depression, but it is insignificant. The factors influencing depression are most likely more diverse and do not depend much on age.
Also, Correlation (Depression vs Have you ever had suicidal thoughts?) has almost 0,55 - which means that people who have had suicidal thoughts are more likely to suffer from depression.

# Hypothesis 1:
# *I assume that men are more prone to depression than women*

```python
import plotly.graph_objects as go
data = df.groupby(['Gender', 'Depression']).size().reset_index(name='Count')
x_labels = ['Male - Have depression', 'Male - No depression',
            'Female - Have depression', 'Female - No depression']
counts = [
    data[(data['Gender'] == 'Male') & (data['Depression'] == 1)]['Count'].values[0] if not data[(data['Gender'] == 'Male') & (data['Depression'] == 1)].empty else 0,
    data[(data['Gender'] == 'Male') & (data['Depression'] == 0)]['Count'].values[0] if not data[(data['Gender'] == 'Male') & (data['Depression'] == 0)].empty else 0,
    data[(data['Gender'] == 'Female') & (data['Depression'] == 1)]['Count'].values[0] if not data[(data['Gender'] == 'Female') & (data['Depression'] == 1)].empty else 0,
    data[(data['Gender'] == 'Female') & (data['Depression'] == 0)]['Count'].values[0] if not data[(data['Gender'] == 'Female') & (data['Depression'] == 0)].empty else 0
]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=x_labels,
    y=counts,
    marker_color=['lightblue', 'lightblue', 'pink', 'pink'],  # Цвет для мужчин и женщин
    text=counts,
    textposition='auto',
    name='Depression Status'
))

fig.update_layout(
    title="Comparison of Gender and Depression",
    xaxis=dict(
        title="Gender and Depression Status",
        tickvals=x_labels
    ),
    yaxis_title="Count",
    paper_bgcolor='white',
    plot_bgcolor='white',
    width=900,
    height=600
)

fig.show()
```

Now we see that most of the respondents, both men and women, suffer from depression. But from past studies, we remember that in general there were more men interviewed, so we will find the percentage of men with depression from the total number of men and the same for women

```python
import plotly.graph_objects as go

male_total = len(df[df['Gender'] == 'Male'])
male_with_depression = len(df[(df['Gender'] == 'Male') & (df['Depression'] == 1)])

female_total = len(df[df['Gender'] == 'Female'])
female_with_depression = len(df[(df['Gender'] == 'Female') & (df['Depression'] == 1)])
male_percentage = (male_with_depression / male_total) * 100 if male_total > 0 else 0

fig_male = go.Figure(go.Pie(
    labels=['Have Depression', 'No Depression'],
    values=[male_with_depression, male_total - male_with_depression],
    marker=dict(colors=['lightblue', 'lightyellow']),
    textinfo='label+percent',
    pull=[0.1, 0]
))

fig_male.update_layout(
    title="Male Depression Percentage",
    paper_bgcolor='white'
)

female_percentage = (female_with_depression / female_total) * 100 if female_total > 0 else 0

fig_female = go.Figure(go.Pie(
    labels=['Have Depression', 'No Depression'],
    values=[female_with_depression, female_total - female_with_depression],
    marker=dict(colors=['pink', 'lightyellow']),
    textinfo='label+percent',
    pull=[0.1, 0]
))

fig_female.update_layout(
    title="Female Depression Percentage",
    paper_bgcolor='white'
)
fig_male.show()
fig_female.show()
```

# Hypothesis 2:
***Depression is a hereditary thing, it is more likely that if there were cases of depression in the family, then the person will also have depression***

```python
import plotly.graph_objects as go

df['Category'] = df.apply(
    lambda row:
    'Family History - no depression, No Depression' if row['Family History of Mental Illness'] == 0 and row['Depression'] == 0 else
    'Family History - had depression, No Depression' if row['Family History of Mental Illness'] == 1 and row['Depression'] == 0 else
    'Family History - no depression, Has Depression' if row['Family History of Mental Illness'] == 0 and row['Depression'] == 1 else
    'Family History - had depression, Has Depression',
    axis=1
)
category_counts = df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']
category_order = [
    'Family History - no depression, No Depression',
    'Family History - had depression, No Depression',
    'Family History - no depression, Has Depression',
    'Family History - had depression, Has Depression'
]
category_counts = category_counts.set_index('Category').reindex(category_order).reset_index()
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=category_counts['Category'],
    y=category_counts['Count'],
    mode='lines+markers',
    line=dict(color='rebeccapurple', width=3),
    marker=dict(size=10, color='orangered'),
    text=category_counts['Count'],
    name='Count'
))
fig.update_layout(
    title='Mental Illness Family History vs Depression Status (Line Chart)',
    xaxis_title='Category',
    yaxis_title='Count',
    xaxis=dict(tickangle=45),
    paper_bgcolor='white',
    plot_bgcolor='whitesmoke',
    height=600,
    width=1000
)

fig.show()
```

# My hypothesis was confirmed
From this graph, it becomes very clear to us that depression is a fairly hereditary thing. A very small number of people who had relatives with depression in their family, but the interviewee did not have it, and a very large number of people who had cases of depression in their family and the interviewee also had depression.Although this information is not one hundred percent reliable and Depression is not a 100% hereditary thing, but it is very likely to be inherited
